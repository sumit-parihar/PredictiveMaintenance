```python
# Load csv file from ./data folder
df = pd.read_csv("../data/machine_data.csv")
df.head()
```

```python
# Extract first letter from Product ID
df['Product Quality'] = df['Product ID'].str[0]
```

```python
# Drop original Product ID , Type(Redundant with Product Quality) , UDI and Failure Type Columns
df = df.drop(columns = ['Product ID', 'Type', 'UDI', 'Failure Type'], errors='ignore')
```

```python
# Convert Product Quality L/M/H into numeric values
le = LabelEncoder()
df['Product Quality'] = le.fit_transform(df['Product Quality'])
# L-> 0 M -> 1 H-> 2
```

```python
# Encode the Target column
df['Target'] = df['Target'].map({'Failure' : 1, 'No Failure' : 0})
```
